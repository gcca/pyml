```python
datos = genfromtxt('data1', delimiter=',')
```

```python
def trazarDatos(x, y):
	plot(x, y, 'rx', markersize=10)
	ylabel('Ganancia ($10 000)')
	xlabel('Población (10 000)')
```

```python
X, y  = datos.T
m = len(y)

trazarDatos(X, y)
```

Datos

```python
X = vstack((ones(m), X)).T
θ = zeros(2)

iteraciones = 1500
α = 0.01
```

Ejecutando *función de costo*

$$ J( \theta ) = \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta }(x^{(i)}) - y^{(i)})^2 $$

$$ h_{\theta}(x) = \theta^{T}x = \theta_0 + \theta{_1}x_1 $$

```python
def calcCosto(X, y, θ):
	m = len(y)

	h = dot(X, θ) - y
	J = dot(h, h)/ (2 * m)

	return J
```

```python
calcCosto(X, y, θ)
```

Ejecutando el *Descenso Gradiente*

\begin{align}
\theta_j & := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta )\\\\
\theta_j & := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_{\theta }(x^{(i)}) - y^{(i)})x_j^{(i)}
\end{align}

```python
def descGradiente(X, y, θ, α, iteraciones):
	m = len(y)
	J_hist = []

	for iter in range(iteraciones):

		θ -= α * dot((dot(X, θ) - y), X) / m

		J_hist.append(calcCosto(X, y, θ))

	return θ, J_hist
```

```python
θ, _ = descGradiente(X, y, θ, α, iteraciones)
```

Theta hallado por el descenso gradiente:

```python
print('%f %f \n' % (θ[0], θ[1]))
```

```python
plot(X[:, 1], y, 'rx', markersize=10, label='Datos de entrenamiento')
plot(X[:, 1], dot(X, θ), '-', label='Regresión lineal')
ylabel('Ganancia ($10 000)')
xlabel('Población (10 000)')
legend()
show()
```

```python
pred1 = dot(array((1, 3.5)), θ)
pred2 = dot(array((1, 7)), θ)

print('Para una población = 35 000, se predice una ganancia de %f\n'
      % (pred1 * 10000))
print('Para una población = 70 000, se predice una ganancia de %f\n'
      % (pred2 * 10000))
```

Visualizando J(θ) ...

Rejilla sobre la cuál se calculará J                                          

```python
θ0_vals = linspace(-10, 10, 100)
θ1_vals = linspace( -1,  4, 100)
```

Fijar J_vals a la matriz de 0's   

```python
J_vals = zeros((θ0_vals.size, θ1_vals.size))

for i in range(θ0_vals.size):
        for j in range(θ1_vals.size):
                t = array((θ0_vals[i], θ1_vals[j]))
                J = calcCosto(X, y, t)
                J_vals[i, j] = J
```

Invertir ejes

```python
J_vals = J_vals.T
```

Trazar superficie

```python
ax = Axes3D(figure())
ax.view_init(20, 230)

θ0_vals, θ1_vals = meshgrid(θ0_vals, θ1_vals)
ax.plot_surface(θ0_vals, θ1_vals, J_vals, cmap = cm.coolwarm, alpha = 0.69)
xlabel('θ₀')
ylabel('θ₁')
show()
```

Trazar contorno

```python
# Trazar J_vals con 15 contornos espaciados logarítmicamente entre 0.01 y 100
figure()
contour(θ0_vals, θ1_vals, J_vals, logspace(-2, 3, 20))
xlabel('θ₀')
ylabel('θ₁')
hold(True)
plot(θ[0], θ[1], 'rx', markersize = 10, markeredgewidth = 2.0)
show()
```
